```python
import math

# 1. 첫 번째 정수 입력받기
while True:
  while True:
      try:
          num1 = input("첫 번째 정수를 입력하세요: ")
          if float(num1) != int(num1):
              raise ValueError
          num1 = int(num1)  # 성공하면 int로 변환
          break  # 성공했으니 다음 단계(num2)로 이동
      except:
          print("잘못된 입력입니다. '정수'를 입력해주세요.")

  # 2. 두 번째 정수 입력받기
  while True:
      try:
          num2 = input("두 번째 정수를 입력하세요: ")
          if float(num2) != int(num2):
              raise ValueError
          num2 = int(num2)
          break # 성공했으니 다음 단계(연산자)로 이동
      except:
          print("잘못된 입력입니다. '정수'를 입력해주세요.")

  # 3. 연산자 입력받기
  while True:
      op = input("연산자를 입력하세요(ex +, -, *, /, **): ")
      if op in ['+', '-', '*', '/', '**']: # 이 연산자만 받을 거니까 if절이 유리함.
          break
      else:
          print("잘못된 연산자입니다. +, -, *, /, ** 중 하나를 입력하세요.")

  #연산하기
  try:
      if op == '+':
          result = num1 + num2
      elif op == '-':
          result = num1 - num2
      elif op == '*':
          result = num1 * num2
      elif op == '/':
          if num2 == 0:
              # 직접 에러를 발생시킵니다.
              raise ZeroDivisionError("0으로 나눌 수 없습니다.")
          result = num1 / num2
      elif op == '**':
          result = math.pow(num1, num2)

      print(f"결과: {num1} {op} {num2} = {result}")

  except ZeroDivisionError as e:
      print(f"계산 오류: {e}")

  #계산기 계속할 것인지 입력받기
  while True:
        choice = input("계속 계산하시겠습니까? 소문자 'y' 또는 'n' 으로만 입력해주세요.: ")

        if choice == 'y':
            print("새로운 계산을 시작합니다.")
            break  # 안쪽 루프를 깨고 나가서 첫 번째 정수 입력으로 돌아감

        elif choice == 'n':
            print("계산기를 종료합니다. 이용해 주셔서 감사합니다!")
            exit() # 프로그램 전체 종료
        else:
            print("알 수 없는 입력입니다. 'y' 또는 'n'을 입력해주세요.")


```
